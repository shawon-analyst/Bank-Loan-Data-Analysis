```python
# Total good loans
good_loans = data[data['loan_status'].isin(['Fully Paid','Current'])]

Good_Loans_Total_Application = good_loans['id'].count()
Good_Loans_Total_Funded_Amount = good_loans['loan_amount'].sum()
Good_Loans_Total_Received_Amount = good_loans['total_payment'].sum()

Good_Loans_Total_Funded_Amount_million = (Good_Loans_Total_Funded_Amount / 1000000).round(2)
Good_Loans_Total_Received_million = (Good_Loans_Total_Received_Amount/ 1000000).round(2)
Good_Loans_Percentage = ((Good_Loans_Total_Application / total_loan_application)*100).round(2)


print(f"GOOD LOAN TOTAL APPLICATION         : {Good_Loans_Total_Application}")
print(f"GOOD LOAN TOTAL FUNDED AMOUNT       : ${Good_Loans_Total_Funded_Amount_million}M")
print(f"GOOD LOAN TOTAL RECEIVED AMOUNT     : ${Good_Loans_Total_Received_million}M")
print(f"PERCENTAGE OF GOOD LOAN APPLICATION : {Good_Loans_Percentage}%")

```

### BAD LOAN METRICS

```python
# Total Bad  loans
BAD_loans = data[data['loan_status'].isin(['Charged Off'])]

BAD_Loans_Total_Application = BAD_loans['id'].count()
BAD_Loans_Total_Funded_Amount = BAD_loans['loan_amount'].sum()
BAD_Loans_Total_Received_Amount = BAD_loans['total_payment'].sum()

BAD_Loans_Total_Funded_Amount_million = (BAD_Loans_Total_Funded_Amount / 1000000).round(2)
BAD_Loans_Total_Received_million = (BAD_Loans_Total_Received_Amount/ 1000000).round(2)
BAD_Loans_Percentage = ((BAD_Loans_Total_Application / total_loan_application)*100).round(2)


print(f"BAD LOAN TOTAL APPLICATION         : {BAD_Loans_Total_Application}")
print(f"BAD LOAN TOTAL FUNDED AMOUNT       : ${BAD_Loans_Total_Funded_Amount_million}M")
print(f"BAD LOAN TOTAL RECEIVED AMOUNT     : ${BAD_Loans_Total_Received_million}M")
print(f"PERCENTAGE OF BAD LOAN APPLICATION : {BAD_Loans_Percentage}%")
```

### 📉 Monthly Loan Trends (Total Funded Amount)

```python
#create month_name and month_num column 

data['month_name'] = data['issue_date'].dt.strftime('%b')
data['month_num'] = data['issue_date'].dt.month

# Loan amount Group by Month_num
Monthly_funded  = data.groupby(["month_num","month_name"])['loan_amount'].sum().reset_index().sort_values('month_name')
Monthly_funded = Monthly_funded.sort_values('month_num').drop('month_num', axis = 1)
Monthly_funded['loan_amount'] = (Monthly_funded['loan_amount'] / 1000000).round(1)

fig = px.line(data_frame= Monthly_funded, x='month_name', y = 'loan_amount',
              text='loan_amount',title='Total Funded Amount by Month',height=450, width=900, template='simple_white')
fig.update_xaxes(showgrid = False)
fig.update_yaxes(showgrid = False)
fig.update_layout(
    title_x = 0.48,
    paper_bgcolor = 'white',
    plot_bgcolor = 'white',
    font = dict(family = 'arial', color = 'black',weight = 'bold')
    
)
fig.update_traces(
    textposition = 'top center',
    line=dict(width=4, color='royalblue'), 
    marker=dict(size=10, symbol='circle'), 
    mode='lines+markers+text',
    textfont = dict(family = 'Arial', weight  ='bold')
)
fig.write_image("Total Funded Amount by Month.png", width=900, height=450, scale=5)
fig.show()
```

## 📉 Monthly Loan Trends (Total Received Amount)

```python

Monthly_recived  = data.groupby(["month_num","month_name"])['total_payment'].sum().reset_index().sort_values('month_name')
Monthly_recived = Monthly_recived.sort_values('month_num').drop('month_num', axis = 1)
Monthly_recived['total_payment'] = (Monthly_recived['total_payment'] / 1000000).round(1)

fig = px.line(data_frame= Monthly_recived, x='month_name', y = 'total_payment',
              text='total_payment',title='Total Received Amount by Month',height=450, width=900, template='simple_white')
fig.update_xaxes(showgrid = False)
fig.update_yaxes(showgrid = False)
fig.update_layout(
    title_x = 0.48,
    paper_bgcolor = 'white',
    plot_bgcolor = 'white',
    font = dict(family = 'arial', color = 'black',weight = 'bold')
    
)
fig.update_traces(
    textposition = 'top center',
    line=dict(width=4, color='green'), 
    marker=dict(size=10, symbol='circle'), 
    mode='lines+markers+text',
    textfont = dict(family = 'Arial', weight  ='bold')
)
fig.show()
```

# 🏦 Bank Loan Analysis - Executive Summary

After a comprehensive analysis of the loan dataset, several key patterns regarding borrower behavior and credit distribution have been identified:

### 1. 🎯 Primary Loan Drivers
* **Debt Consolidation:** The vast majority of loan applications are driven by 'Debt Consolidation.' This indicates that a significant portion of the customer base is seeking to manage or refinance existing liabilities. 💰
* **Secondary Needs:** Credit card payments and home improvements emerge as the next most frequent reasons for borrowing, highlighting a focus on personal financial restructuring. 💳

### 2. 🏠 Housing & Risk Correlation
* **Renters and Mortgage Holders:** Borrowers who live in rented accommodations or have existing mortgages account for the highest total funded amounts. 🏠
* **Homeowners:** Those who fully own their homes represent the smallest segment of borrowers, which may suggest higher financial stability and lower credit dependency within this group. 💸

### 3. 💼 Impact of Employment Stability
* **Job Tenure:** There is a direct correlation between employment length and the total funded amount. Borrowers with **10+ years** of experience receive the highest volume of funding, suggesting that the bank favors long-term employment as a key indicator of repayment capacity. 📈

### 4. 📍 Geographical Distribution
* **Key Regions:** Loan activity is heavily concentrated in specific states (e.g., CA, NY, TX). These regions represent the highest market share, and targeted marketing strategies in these areas could further optimize loan growth. 🗺️

### 5. 📉 Seasonal Trends
* **Quarterly Growth:** The analysis shows a gradual increase in loan funding toward the end of the calendar year. This seasonal trend is likely linked to holiday spending and year-end financial planning by customers. ⏳
